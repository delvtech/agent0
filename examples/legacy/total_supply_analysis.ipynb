```python
import numpy as np
import math 
import pandas as pd
from pandas.io.json import json_normalize
import json
import os, sys
root_dir = os.path.join(os.path.join(os.getcwd(), os.pardir), os.pardir)
sys.path.insert(1, root_dir)
from sim import *

# initialize pool
x_reserves = 0
y_reserves = 0
x_in = 1
y_in = 0
total_supply = 0
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
print("Initial Deposit: x_needed={:}, y_needed={:}, lp_out={:}".format(x_needed, y_needed, total_supply) )

# calculate what trade to make to set APY
x_reserves += x_in
y_reserves += y_in
total_supply += lp_out
APY = .01
ts = ElementPricingModel().calc_time_stretch(APY*100)
print("For APY = {:}%: time_stretch={:}".format(APY*100,ts))
days_until_maturity=180
t=days_until_maturity/(365*ts)
T=days_until_maturity/365
y_needed = x_reserves * ((1-APY*T)**(1/-t)-1)/(1+(1-APY*T)**(1/-t))
print("For APY = {:}%: y_needed={:}".format(APY*100,y_needed) )

(without_fee_or_slippage,with_fee,without_fee,fee) = ElementPricingModel().calc_out_given_in(y_needed,y_reserves+total_supply,x_reserves,"base",.1,t)
x_reserves -= with_fee
y_reserves += y_needed
price = ElementPricingModel().calc_spot_price(x_reserves,y_reserves,total_supply,t)
APY = ElementPricingModel().apy(price,days_until_maturity)
print("Resulting APY= {:}%".format(APY) )

x_in = 1
y_in = 1
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
x_in = x_needed
y_in = y_needed
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
print("Deposit: x_needed={:}, y_needed={:} for lp shares={:}".format(x_needed, y_needed, lp_out))
x_reserves += x_in
y_reserves += y_in
total_supply+=lp_out
print("x_reserves={:}, y_reserves={:}, total_supply={:} estimate_total_supply={:}".format(x_reserves, y_reserves, total_supply, x_reserves+price*y_reserves) )

x_in = 1
y_in = 1
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
x_in = x_needed
y_in = y_needed
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
print("Deposit: x_needed={:}, y_needed={:} for lp shares={:}".format(x_needed, y_needed, lp_out))
x_reserves += x_in
y_reserves += y_in
total_supply+=lp_out
print("x_reserves={:}, y_reserves={:}, total_supply={:} estimate_total_supply={:}".format(x_reserves, y_reserves, total_supply, x_reserves+price*y_reserves) )

x_in = 1
y_in = 1
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
x_in = x_needed
y_in = y_needed
(x_needed, y_needed, lp_out) = ElementPricingModel().calc_lp_out_given_tokens_in(x_in, y_in, x_reserves, y_reserves, total_supply)
print("Deposit: x_needed={:}, y_needed={:} for lp shares={:}".format(x_needed, y_needed, lp_out))
x_reserves += x_in
y_reserves += y_in
total_supply+=lp_out
print("x_reserves={:}, y_reserves={:}, total_supply={:} estimate_total_supply={:}".format(x_reserves, y_reserves, total_supply, x_reserves+price*y_reserves) )
```
